# How does [company] practice openness? 

```python
## Cohere 


cohere_df = df[df["display_org"] == "Cohere"]

for license_type, group in cohere_df.groupby("license"):
    print(f"\n=== License: {license_type} ({len(group)} models) ===")
    for model in group["model_id"]:
        print(f" - {model}")


```

```python
## Mistral
mistral_df = df[df["display_org"] == "Mistral"]

for license_type, group in mistral_df.groupby("license"):
    print(f"\n=== License: {license_type} ({len(group)} models) ===")
    for model, url in zip(group["model_id"], group["url"]):
        print(f" - {model} → {url}")
```

```python
openai_df = df[df["display_org"] == "OpenAI"]

for license_type, group in openai_df.groupby("license", dropna=False):
    print(f"\n=== License: {license_type if pd.notna(license_type) else 'None'} ({len(group)} models) ===")
    for model in group["model_id"]:
        print(f" - {model}")
```

```python
## META

meta_df = df[df["display_org"] == "Meta (Llama)"]

for license_type, group in meta_df.groupby("license", dropna=False):
    print(f"\n=== License: {license_type if pd.notna(license_type) else 'None'} ({len(group)} models) ===")
    for model in group["model_id"]:
        print(f" - {model}")
```

```python
# 1) Collapse Google models into parent "families" and show the top ones
google_df = df[df["display_org"] == "Google"].copy()
google_df["repo"] = google_df["model_id"].str.split("/", n=1).str[1]
google_df["family"] = (
    google_df["repo"]
    .str.replace(r"^([A-Za-z]+).*", r"\1", regex=True)  # take leading letters as family
    .str.lower()
)

print("=== Top Google model families (parent categories) ===")
print(google_df["family"].value_counts().head(30))

# 2) Focus on Gemma: list models grouped by license (include null)
gemma_df = google_df[google_df["family"] == "gemma"]

for license_type, group in gemma_df.groupby("license", dropna=False):
    label = license_type if pd.notna(license_type) else "None"
    print(f"\n=== Gemma — License: {label} ({len(group)} models) ===")
    for model in group["model_id"]:
        print(f" - {model}")
```
